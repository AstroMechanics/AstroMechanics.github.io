```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import eval_legendre
plt.style.use("dynamics_site")
```

## System Setup

```python
r1 = 1.0
r2 = 1.5**(2/3)
p1 = r1**(3/2)
p2 = r2**(3/2)
```

```python
tgrid = np.linspace(0,7,1000)
x1grid = r1*np.sin(2*np.pi*tgrid/p1)
y1grid = r1*np.cos(2*np.pi*tgrid/p1)
x2grid = r2*np.sin(2*np.pi*tgrid/p2)
y2grid = r2*np.cos(2*np.pi*tgrid/p2)

theta1 = np.arctan2(x1grid,y1grid)
theta2 = np.arctan2(x2grid,y2grid)
```

```python
fig,ax = plt.subplots(ncols=2,figsize=(15,7))
ax[0].scatter(x1grid,y1grid,label=r'planet 1')
ax[0].scatter(x2grid,y2grid,label=r'planet 2')
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].legend()
ax[1].plot(tgrid,theta1,label=r'$\theta_1$')
ax[1].plot(tgrid,theta2,label=r'$\theta_2$')
ax[1].set_xlabel("t")
ax[1].set_ylabel(r"$\theta$")
ax[1].legend()
plt.savefig('legendre1.png')
```

```python
theta12 = theta1-theta2
h = r1/r2
phi = (1/r2)*1/np.sqrt(1**2+h**2 - 2*h*np.cos(theta12))

res = lambda n: np.sum([(h**l)*eval_legendre(l,np.cos(theta12)) for l in range(0,n)],axis=0) 
```

```python
fig,ax = plt.subplots(figsize=(15,7))
plt.plot(tgrid,phi,c='k',label=r'$\phi$',lw=10)
vals = [2,3,5,10,50]
for v in vals:
    plt.plot(tgrid,(1/r2)*res(v),label='n='+str(v))
plt.ylim(0,4)
plt.legend(loc='best')
plt.xlabel(r'$time$')
plt.ylabel(r'$\phi$')
plt.savefig('legendre2.png')
```
