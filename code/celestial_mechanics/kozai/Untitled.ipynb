```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dynamics_site')
from scipy.integrate import odeint
```

```python
omega = np.linspace(0*np.pi,2*np.pi,1000)
egrid = np.linspace(0,1.0,1000)
O,E = np.meshgrid(omega,egrid)

Jz = 0.1
num = E**4 + 2*Jz**2 + (E**2)*(-1+3*Jz**2) - 5*(E**2)*(-1 + E**2 + Jz**2)*np.cos(2*O)
den = 2*(-1+E**2)
z = num/den
```

```python
sc=plt.imshow(z,extent=[-3*np.pi,3*np.pi,-2*np.pi,2*np.pi])
plt.close()
fig = plt.subplots(figsize=(15,7))
plt.contour(O,E,z,levels=75)
plt.xlabel(r"$\omega$")
plt.ylabel(r"$e_1$")
plt.colorbar(sc,label='Energy')
plt.contour(O,E,z,levels=[10],colors='k',linewidths=5,linestyles='dashed')
```

## Now \theta , \omega plane

```python
omega = np.linspace(0,2*np.pi,1000)
theta = np.linspace(0.0,1.0,1000)
O,T = np.meshgrid(omega,theta)

Jz = 0.2
num = Jz**2 - (1+3*Jz**2)*(T**2) + 5*(T**4) + 5*(Jz-T)*(Jz+T)*(-1+T**2)*np.cos(2*O)
den = 2*(T**2)
z = num/den
```

```python
sc=plt.imshow(z,extent=[-3*np.pi,3*np.pi,-2*np.pi,2*np.pi],vmax=2.5,vmin=-2.5)
plt.close()
fig = plt.subplots(figsize=(15,7))
plt.contour(O,E,z,levels=np.linspace(-2.5,2.5,20))
plt.xlabel(r"$\omega$")
plt.ylabel(r"$\theta$")
plt.colorbar(sc,label='H')
#plt.contour(O,E,z,levels=[10],colors='k',linewidths=5,linestyles='dashed')
```
