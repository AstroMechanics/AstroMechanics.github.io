```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dynamics_site')
```

```python
m=1
l=1
g=10
```

```python
theta = np.linspace(-3*np.pi,3*np.pi,1000)
Ptheta = np.linspace(-3*np.pi,3*np.pi,1000)
T,PT = np.meshgrid(theta,Ptheta)
z = PT**2/(2*m*l**2) + m*g*l*(1-np.cos(T))
```

```python
sc=plt.imshow(z,extent=[-3*np.pi,3*np.pi,-2*np.pi,2*np.pi])
plt.close()
fig = plt.subplots(figsize=(15,7))
plt.contour(T,PT,z,levels=10)
plt.xlabel(r"$\theta$")
plt.ylabel(r"$P_\theta$")
plt.colorbar(sc,label='Energy')
plt.contour(T,PT,z,levels=[g*l*m-1e1,g*l*m+1e1],colors='k',linewidths=5,linestyles='dashed')
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-3*np.pi,3*np.pi)
plt.savefig('hamiltonian_pendulum.png')
```
